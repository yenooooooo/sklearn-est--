### 선형 회귀 모델

### 데이터셋을 생성해서 준비

```python
import numpy as np 

# 선형회귀를 위한 데이터셋을 직접 만들어보자
# y = 2x+1
# x = [1,2,3,4,5]
# y = [3,5,7,9,11]

n = 10
x_total = np.arange(1, n+1).reshape(-1,1) 
x_total
```

```python
# y = 2*x+1 + 난수(0~3사이의 난수 10개)
randValue = (np.random.random(n)*7).reshape(-1,1)
y_total = 2*x_total+1 + randValue
y_total
```

```python
# x_total, y_toal을 산점도 그래프로 그리기
import matplotlib.pyplot as plt
plt.scatter(x_total, y_total)
plt.show()
```

### 데이터셋을 나누기(x_total, y_total)

```python
# x_total, y_total을 나누기
from sklearn.model_selection import train_test_split

# 나누는 비율 0.8:0.2
x_train, x_test, y_train, y_test = train_test_split(x_total, y_total, test_size=0.2, random_state=42)
```

```python
x_train, x_test
```

### 선형회귀 모델을 사용해서 학습, 평가

```python
# sklearn의 선형회귀 모델을 사용
from sklearn.linear_model import LinearRegression

# 1. 모델 생성(선언)
model = LinearRegression()

# 2. 모델 학습
model.fit(x_train, y_train)

# 3. 모델 예측
y_pred = model.predict(x_test)

# 4. 모델 평가
from sklearn.metrics import mean_squared_error, mean_absolute_error, root_mean_squared_error
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
print(f"Root Mean Squared Error: {rmse}")
```
