```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import pandas as pd

# 1. Iris    DataFrame 
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target

# 2.    PCA
fig = plt.figure(figsize=(8, 6))

#  : Axes3D    add_subplot 
ax = fig.add_subplot(111, projection='3d', elev=-150, azim=110)

X_reduced = PCA(n_components=3).fit_transform(df.drop('target', axis=1))

# 3.  
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=df['target'],
           cmap=plt.cm.Set1, edgecolor='k', s=40)

ax.set_title("Iris 3D PCA")
ax.set_xlabel("1st")
ax.xaxis.set_ticklabels([])
ax.set_ylabel("2nd")
ax.yaxis.set_ticklabels([])
ax.set_zlabel("3rd")
ax.zaxis.set_ticklabels([])

plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import pandas as pd

# 1.  
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target

# 2.   (  X_train )
#  ! (   X,   y )
X_train, X_test, y_train, y_test = train_test_split(
    df.drop('target', axis=1), 
    df['target'], 
    test_size=0.2, 
    random_state=8
)
```

```python
#   CPU  
import os
import multiprocessing

#  1: os   ( )
#  ()      .
n_cores_os = os.cpu_count()
print(f"CPU   (os.cpu_count): {n_cores_os}")

#  2: multiprocessing  
n_cores_mp = multiprocessing.cpu_count()
print(f"CPU   (multiprocessing.cpu_count): {n_cores_mp}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# 1.   ( )
# random_state    .
model = RandomForestClassifier(random_state=42)

# 2.    
# n_estimators: 10, 20, 30, 40, 50
params = {
    'n_estimators': range(10, 51, 10)
}

# 3. GridSearchCV  
# cv=5: 5-Fold   
# refit=True:           
# sklearn n_jobs=-1 :  CPU    
grid_cv = GridSearchCV(model, param_grid=params, cv=5, n_jobs=-1, scoring='accuracy')

# 4.   (Grid Search )
grid_cv.fit(X_train, y_train)

# 5.     
print(f" : {grid_cv.best_params_}")
print(f"  : {grid_cv.best_score_:.4f}")

# 6.       
best_rf = grid_cv.best_estimator_
y_pred = best_rf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"   (Best Estimator): {accuracy:.4f}")
```
