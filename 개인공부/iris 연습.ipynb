```python
import numpy as np 

# 선형회귀를 위한 데이터셋을 직접 만들어보자
# 목표 식: y = 2x + 1
# x = [1,2,3,4,5] 일 때 y = [3,5,7,9,11] 이 나오도록

n = 10
# 1부터 n(10)까지 순서대로 생성하고, 계산을 위해 (10행, 1열) 형태의 2차원 배열로 변환
x_total = np.arange(1, n+1).reshape(-1,1) 
x_total # 생성된 x 입력 데이터 확인
```

```python
# y = 2*x + 1 식에 약간의 무작위 값(노이즈)을 더해 현실적인 데이터처럼 만듦
# 0~1 사이의 난수를 만들고 7을 곱해 노이즈 크기를 키움
randValue = (np.random.random(n)*7).reshape(-1,1)

# 깨끗한 수식 결과에 노이즈를 더함
y_total = 2*x_total+1 + randValue
y_total # 생성된 y 정답 데이터 확인
```

```python
# x_total, y_total 데이터를 눈으로 확인하기 위해 산점도(점 그래프) 그리기
import matplotlib.pyplot as plt

# x축에는 입력값, y축에는 정답값을 찍어서 분포 확인
plt.scatter(x_total, y_total)
plt.show() # 그래프 출력
```

```python
# 전체 데이터를 학습용(Train)과 테스트용(Test)으로 나누기
from sklearn.model_selection import train_test_split

# test_size=0.2: 전체의 20%는 테스트(채점)용으로 따로 빼두겠다는 의미
# random_state=42: 실행할 때마다 데이터가 다르게 섞이지 않도록 난수 시드 고정
x_train, x_test, y_train, y_test = train_test_split(x_total, y_total, test_size=0.2, random_state=42)

# 잘 나뉘었는지 데이터 확인
x_train, x_test 
```

```python
# sklearn 라이브러리에서 선형회귀 모델 가져오기
from sklearn.linear_model import LinearRegression

# 1. 모델 생성(선언): "나는 선형 회귀라는 방법을 쓸 거야" 라고 객체 생성
model = LinearRegression()

# 2. 모델 학습(Fit): "이 문제(x_train)와 정답(y_train)을 줄 테니 공부해서 규칙을 찾아봐"
model.fit(x_train, y_train)

# 3. 모델 예측(Predict): "공부가 끝났으니 안 보여줬던 문제(x_test)를 풀어봐"
y_pred = model.predict(x_test)

# 4. 모델 평가하기 (제대로 맞췄는지 채점)
from sklearn.metrics import mean_squared_error, mean_absolute_error, root_mean_squared_error

# MSE(Mean Squared Error): 틀린 오차를 제곱해서 평균 낸 것 (작을수록 좋음)
mse = mean_squared_error(y_test, y_pred)
# MAE(Mean Absolute Error): 틀린 오차의 절대값 평균 (직관적으로 얼마 틀렸는지 알기 쉬움)
mae = mean_absolute_error(y_test, y_pred)
# RMSE(Root MSE): MSE에 루트를 씌워 원래 단위와 비슷하게 만든 것
rmse = root_mean_squared_error(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```
